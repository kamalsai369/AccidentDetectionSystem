```python
# Accident Detection System using TensorFlow
# Complete implementation with data preprocessing, model training, and prediction

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

class AccidentDetectionSystem:
    def __init__(self, data_dir='accident_dataset', img_size=(224, 224), batch_size=32):
        self.data_dir = data_dir
        self.img_size = img_size
        self.batch_size = batch_size
        self.model = None
        self.history = None
        self.class_names = ['accident', 'no_accident']  # Adjust based on your classes

    def setup_data_generators(self):
        """Setup data generators for training, validation, and testing"""
        # Data augmentation for training set
        train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode='nearest'
        )

        # Only rescaling for validation and test sets
        val_test_datagen = ImageDataGenerator(rescale=1./255)

        # Create generators
        self.train_generator = train_datagen.flow_from_directory(
            os.path.join(self.data_dir, 'data', 'train'),
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            shuffle=True
        )

        self.val_generator = val_test_datagen.flow_from_directory(
            os.path.join(self.data_dir, 'data', 'val'),
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            shuffle=False
        )

        self.test_generator = val_test_datagen.flow_from_directory(
            os.path.join(self.data_dir, 'data', 'test'),
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            shuffle=False
        )

        # Get class names from generator
        self.class_names = list(self.train_generator.class_indices.keys())
        self.num_classes = len(self.class_names)

        print(f"Found {self.train_generator.samples} training images")
        print(f"Found {self.val_generator.samples} validation images")
        print(f"Found {self.test_generator.samples} test images")
        print(f"Classes: {self.class_names}")

    def create_cnn_model(self):
        """Create a custom CNN model for accident detection"""
        model = keras.Sequential([
            # First Convolutional Block
            layers.Conv2D(32, (3, 3), activation='relu', input_shape=(*self.img_size, 3)),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.25),

            # Second Convolutional Block
            layers.Conv2D(64, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.25),

            # Third Convolutional Block
            layers.Conv2D(128, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.25),

            # Fourth Convolutional Block
            layers.Conv2D(256, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.25),

            # Global Average Pooling
            layers.GlobalAveragePooling2D(),

            # Dense layers
            layers.Dense(512, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),

            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),

            # Output layer
            layers.Dense(self.num_classes, activation='softmax')
        ])

        return model

    def create_transfer_learning_model(self, base_model_name='MobileNetV2'):
        """Create a transfer learning model using pre-trained weights"""
        # Load pre-trained base model
        if base_model_name == 'MobileNetV2':
            base_model = keras.applications.MobileNetV2(
                input_shape=(*self.img_size, 3),
                include_top=False,
                weights='imagenet'
            )
        elif base_model_name == 'ResNet50':
            base_model = keras.applications.ResNet50(
                input_shape=(*self.img_size, 3),
                include_top=False,
                weights='imagenet'
            )
        elif base_model_name == 'EfficientNetB0':
            base_model = keras.applications.EfficientNetB0(
                input_shape=(*self.img_size, 3),
                include_top=False,
                weights='imagenet'
            )

        # Freeze base model layers
        base_model.trainable = False

        # Add custom classification head
        model = keras.Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(self.num_classes, activation='softmax')
        ])

        return model

    def compile_and_train(self, model_type='transfer', base_model='MobileNetV2', epochs=50):
        """Compile and train the model"""
        # Create model
        if model_type == 'cnn':
            self.model = self.create_cnn_model()
        else:
            self.model = self.create_transfer_learning_model(base_model)

        # Compile model
        self.model.compile(
            optimizer=keras.optimizers.Adam(learning_rate=0.001),
            loss='categorical_crossentropy',
            metrics=['accuracy', 'precision', 'recall']
        )

        print(self.model.summary())

        # Define callbacks
        callbacks = [
            EarlyStopping(
                monitor='val_accuracy',
                patience=10,
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=5,
                min_lr=1e-7,
                verbose=1
            ),
            ModelCheckpoint(
                'best_accident_model.h5',
                monitor='val_accuracy',
                save_best_only=True,
                verbose=1
            )
        ]

        # Train the model
        self.history = self.model.fit(
            self.train_generator,
            steps_per_epoch=self.train_generator.samples // self.batch_size,
            epochs=epochs,
            validation_data=self.val_generator,
            validation_steps=self.val_generator.samples // self.batch_size,
            callbacks=callbacks,
            verbose=1
        )

        return self.history

    def fine_tune_model(self, epochs=20):
        """Fine-tune the pre-trained model by unfreezing some layers"""
        if self.model is None:
            print("Please train the model first!")
            return

        # Unfreeze the base model
        base_model = self.model.layers[0]
        base_model.trainable = True

        # Fine-tune from this layer onwards
        fine_tune_at = len(base_model.layers) // 2

        # Freeze all the layers before fine_tune_at
        for layer in base_model.layers[:fine_tune_at]:
            layer.trainable = False

        # Recompile with lower learning rate
        self.model.compile(
            optimizer=keras.optimizers.Adam(learning_rate=0.0001/10),
            loss='categorical_crossentropy',
            metrics=['accuracy', 'precision', 'recall']
        )

        # Continue training
        fine_tune_history = self.model.fit(
            self.train_generator,
            steps_per_epoch=self.train_generator.samples // self.batch_size,
            epochs=epochs,
            validation_data=self.val_generator,
            validation_steps=self.val_generator.samples // self.batch_size,
            callbacks=[
                EarlyStopping(patience=5, restore_best_weights=True),
                ReduceLROnPlateau(factor=0.5, patience=3, min_lr=1e-8)
            ],
            verbose=1
        )

        return fine_tune_history

    def evaluate_model(self):
        """Evaluate the model on test data"""
        if self.model is None:
            print("Please train the model first!")
            return

        # Evaluate on test set
        test_loss, test_accuracy, test_precision, test_recall = self.model.evaluate(
            self.test_generator,
            steps=self.test_generator.samples // self.batch_size
        )

        print(f"\nTest Results:")
        print(f"Test Accuracy: {test_accuracy:.4f}")
        print(f"Test Precision: {test_precision:.4f}")
        print(f"Test Recall: {test_recall:.4f}")
        print(f"Test F1-Score: {2 * (test_precision * test_recall) / (test_precision + test_recall):.4f}")

        # Generate predictions for confusion matrix
        self.test_generator.reset()
        predictions = self.model.predict(
            self.test_generator,
            steps=self.test_generator.samples // self.batch_size
        )

        y_pred = np.argmax(predictions, axis=1)
        y_true = self.test_generator.classes[:len(y_pred)]

        # Classification report
        print("\nClassification Report:")
        print(classification_report(y_true, y_pred, target_names=self.class_names))

        # Confusion Matrix
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.class_names, yticklabels=self.class_names)
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()

        return test_accuracy, test_precision, test_recall

    def plot_training_history(self):
        """Plot training history"""
        if self.history is None:
            print("No training history available!")
            return

        fig, axes = plt.subplots(2, 2, figsize=(15, 10))

        # Plot accuracy
        axes[0, 0].plot(self.history.history['accuracy'], label='Training Accuracy')
        axes[0, 0].plot(self.history.history['val_accuracy'], label='Validation Accuracy')
        axes[0, 0].set_title('Model Accuracy')
        axes[0, 0].set_xlabel('Epoch')
        axes[0, 0].set_ylabel('Accuracy')
        axes[0, 0].legend()

        # Plot loss
        axes[0, 1].plot(self.history.history['loss'], label='Training Loss')
        axes[0, 1].plot(self.history.history['val_loss'], label='Validation Loss')
        axes[0, 1].set_title('Model Loss')
        axes[0, 1].set_xlabel('Epoch')
        axes[0, 1].set_ylabel('Loss')
        axes[0, 1].legend()

        # Plot precision
        axes[1, 0].plot(self.history.history['precision'], label='Training Precision')
        axes[1, 0].plot(self.history.history['val_precision'], label='Validation Precision')
        axes[1, 0].set_title('Model Precision')
        axes[1, 0].set_xlabel('Epoch')
        axes[1, 0].set_ylabel('Precision')
        axes[1, 0].legend()

        # Plot recall
        axes[1, 1].plot(self.history.history['recall'], label='Training Recall')
        axes[1, 1].plot(self.history.history['val_recall'], label='Validation Recall')
        axes[1, 1].set_title('Model Recall')
        axes[1, 1].set_xlabel('Epoch')
        axes[1, 1].set_ylabel('Recall')
        axes[1, 1].legend()

        plt.tight_layout()
        plt.show()

    def predict_single_image(self, image_path, show_image=True):
        """Predict accident for a single image"""
        if self.model is None:
            print("Please train the model first!")
            return None

        # Load and preprocess image
        img = cv2.imread(image_path)
        if img is None:
            print(f"Could not load image: {image_path}")
            return None

        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, self.img_size)
        img = img.astype('float32') / 255.0
        img = np.expand_dims(img, axis=0)

        # Make prediction
        prediction = self.model.predict(img)
        predicted_class = np.argmax(prediction)
        confidence = prediction[0][predicted_class]

        result = {
            'predicted_class': self.class_names[predicted_class],
            'confidence': confidence,
            'all_predictions': {self.class_names[i]: prediction[0][i] for i in range(len(self.class_names))}
        }

        if show_image:
            plt.figure(figsize=(8, 6))
            img_display = cv2.imread(image_path)
            img_display = cv2.cvtColor(img_display, cv2.COLOR_BGR2RGB)
            plt.imshow(img_display)
            plt.title(f"Predicted: {result['predicted_class']} ({confidence:.2%} confidence)")
            plt.axis('off')
            plt.show()

        return result

    def save_model(self, filepath='accident_detection_model.h5'):
        """Save the trained model"""
        if self.model is None:
            print("No model to save!")
            return

        self.model.save(filepath)
        print(f"Model saved to {filepath}")

    def load_model(self, filepath='accident_detection_model.h5'):
        """Load a trained model"""
        try:
            self.model = keras.models.load_model(filepath)
            print(f"Model loaded from {filepath}")
        except Exception as e:
            print(f"Error loading model: {e}")

# Real-time prediction class for video streams
class RealTimeAccidentDetector:
    def __init__(self, model_path, img_size=(224, 224), class_names=['accident', 'no_accident']):
        self.model = keras.models.load_model(model_path)
        self.img_size = img_size
        self.class_names = class_names

    def detect_from_webcam(self, camera_index=0, confidence_threshold=0.7):
        """Detect accidents from webcam feed"""
        cap = cv2.VideoCapture(camera_index)

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Preprocess frame
            img = cv2.resize(frame, self.img_size)
            img = img.astype('float32') / 255.0
            img = np.expand_dims(img, axis=0)

            # Predict
            prediction = self.model.predict(img, verbose=0)
            predicted_class = np.argmax(prediction)
            confidence = prediction[0][predicted_class]

            # Display results
            label = f"{self.class_names[predicted_class]}: {confidence:.2%}"
            color = (0, 0, 255) if self.class_names[predicted_class] == 'accident' and confidence > confidence_threshold else (0, 255, 0)

            cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

            if self.class_names[predicted_class] == 'accident' and confidence > confidence_threshold:
                cv2.putText(frame, "ACCIDENT DETECTED!", (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

            cv2.imshow('Accident Detection', frame)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()

# Main execution
if __name__ == "__main__":
    # Initialize the accident detection system
    detector = AccidentDetectionSystem(
        data_dir='accident_dataset',
        img_size=(224, 224),
        batch_size=32
    )

    print("Setting up data generators...")
    detector.setup_data_generators()

    print("\nTraining the model...")
    # Train with transfer learning (recommended)
    history = detector.compile_and_train(
        model_type='transfer',
        base_model='MobileNetV2',
        epochs=30
    )

    # Optional: Fine-tune the model
    print("\nFine-tuning the model...")
    detector.fine_tune_model(epochs=10)

    print("\nEvaluating the model...")
    detector.evaluate_model()

    print("\nPlotting training history...")
    detector.plot_training_history()

    print("\nSaving the model...")
    detector.save_model('final_accident_detection_model.h5')

    # Example of single image prediction
    # detector.predict_single_image('path/to/test/image.jpg')

    print("\nTraining completed! You can now use the model for predictions.")

    # For real-time detection (uncomment to use)
    # real_time_detector = RealTimeAccidentDetector('final_accident_detection_model.h5')
    # real_time_detector.detect_from_webcam()
```

```python
# Accident Detection Prediction System
# Complete prediction module for various input types

import tensorflow as tf
from tensorflow import keras
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import json
from datetime import datetime
import time
from pathlib import Path
import pandas as pd
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

class AccidentPredictor:
    def __init__(self, model_path, img_size=(224, 224), class_names=None, confidence_threshold=0.7):
        """
        Initialize the Accident Predictor

        Args:
            model_path (str): Path to the trained model
            img_size (tuple): Input image size (width, height)
            class_names (list): List of class names
            confidence_threshold (float): Minimum confidence for positive prediction
        """
        self.model_path = model_path
        self.img_size = img_size
        self.confidence_threshold = confidence_threshold

        # Load the model
        try:
            self.model = keras.models.load_model(model_path)
            print(f" Model loaded successfully from {model_path}")
        except Exception as e:
            raise Exception(f"Error loading model: {e}")

        # Set class names
        if class_names is None:
            self.class_names = ['accident', 'no_accident']
        else:
            self.class_names = class_names

        print(f" Predictor initialized with classes: {self.class_names}")
        print(f" Confidence threshold: {confidence_threshold}")

    def preprocess_image(self, image_input, source_type='path'):
        """
        Preprocess image for prediction

        Args:
            image_input: Image path, numpy array, or PIL image
            source_type: 'path', 'array', or 'pil'

        Returns:
            preprocessed_image: Ready for model prediction
        """
        try:
            if source_type == 'path':
                # Load from file path
                image = cv2.imread(image_input)
                if image is None:
                    raise ValueError(f"Could not load image from {image_input}")
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            elif source_type == 'array':
                # Already a numpy array
                image = image_input
                if len(image.shape) == 3 and image.shape[2] == 3:
                    # Assume BGR, convert to RGB
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            elif source_type == 'pil':
                # PIL Image
                image = np.array(image_input)

            # Resize image
            image = cv2.resize(image, self.img_size)

            # Normalize
            image = image.astype('float32') / 255.0

            # Add batch dimension
            image = np.expand_dims(image, axis=0)

            return image

        except Exception as e:
            raise Exception(f"Error preprocessing image: {e}")

    def predict_single_image(self, image_path, show_result=True, save_result=False, output_dir='predictions'):
        """
        Predict accident for a single image

        Args:
            image_path (str): Path to the image
            show_result (bool): Display the result
            save_result (bool): Save prediction result
            output_dir (str): Directory to save results

        Returns:
            dict: Prediction results
        """
        try:
            # Preprocess image
            processed_image = self.preprocess_image(image_path, 'path')

            # Make prediction
            start_time = time.time()
            predictions = self.model.predict(processed_image, verbose=0)
            prediction_time = time.time() - start_time

            # Get results
            predicted_class_idx = np.argmax(predictions[0])
            confidence = predictions[0][predicted_class_idx]
            predicted_class = self.class_names[predicted_class_idx]

            # Create result dictionary
            result = {
                'image_path': image_path,
                'predicted_class': predicted_class,
                'confidence': float(confidence),
                'is_accident': predicted_class == 'accident' and confidence >= self.confidence_threshold,
                'prediction_time': prediction_time,
                'timestamp': datetime.now().isoformat(),
                'all_probabilities': {
                    self.class_names[i]: float(predictions[0][i])
                    for i in range(len(self.class_names))
                }
            }

            # Display result
            if show_result:
                self._display_single_prediction(image_path, result)

            # Save result
            if save_result:
                self._save_prediction_result(result, output_dir)

            return result

        except Exception as e:
            print(f"Error predicting image {image_path}: {e}")
            return None

    def predict_batch_images(self, image_folder, save_results=True, output_dir='batch_predictions'):
        """
        Predict accidents for multiple images in a folder

        Args:
            image_folder (str): Path to folder containing images
            save_results (bool): Save batch results
            output_dir (str): Directory to save results

        Returns:
            list: List of prediction results
        """
        print(f" Starting batch prediction for images in: {image_folder}")

        # Get all image files
        image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']
        image_paths = []

        for ext in image_extensions:
            image_paths.extend(Path(image_folder).glob(f'*{ext}'))
            image_paths.extend(Path(image_folder).glob(f'*{ext.upper()}'))

        if not image_paths:
            print(" No images found in the specified folder")
            return []

        print(f" Found {len(image_paths)} images to process")

        results = []
        accident_count = 0

        # Process each image
        for i, image_path in enumerate(image_paths):
            print(f" Processing {i+1}/{len(image_paths)}: {image_path.name}")

            result = self.predict_single_image(str(image_path), show_result=False, save_result=False)

            if result:
                results.append(result)
                if result['is_accident']:
                    accident_count += 1
                    print(f" ACCIDENT DETECTED in {image_path.name} (Confidence: {result['confidence']:.2%})")
                else:
                    print(f" No accident in {image_path.name} (Confidence: {result['confidence']:.2%})")

        # Summary
        print(f"\n Batch Prediction Summary:")
        print(f"   Total images processed: {len(results)}")
        print(f"   Accidents detected: {accident_count}")
        print(f"   No accidents: {len(results) - accident_count}")

        # Save results
        if save_results:
            self._save_batch_results(results, output_dir)

        return results

    def predict_video(self, video_path, output_path=None, show_live=True, save_frames=False):
        """
        Predict accidents in video file

        Args:
            video_path (str): Path to video file
            output_path (str): Path to save output video
            show_live (bool): Show live prediction
            save_frames (bool): Save accident frames

        Returns:
            dict: Video prediction results
        """
        print(f" Starting video prediction for: {video_path}")

        # Open video
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            raise ValueError(f"Could not open video: {video_path}")

        # Get video properties
        fps = int(cap.get(cv2.CAP_PROP_FPS))
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        print(f" Video info: {width}x{height}, {fps} FPS, {total_frames} frames")

        # Setup video writer if saving output
        if output_path:
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        # Initialize tracking variables
        results = {
            'video_path': video_path,
            'total_frames': total_frames,
            'accident_frames': [],
            'predictions': [],
            'summary': {
                'accidents_detected': 0,
                'accident_periods': [],
                'highest_confidence': 0.0
            }
        }

        frame_number = 0
        accident_start = None

        print(" Processing video frames...")

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Make prediction
            processed_frame = self.preprocess_image(frame, 'array')
            prediction = self.model.predict(processed_frame, verbose=0)

            predicted_class_idx = np.argmax(prediction[0])
            confidence = prediction[0][predicted_class_idx]
            predicted_class = self.class_names[predicted_class_idx]
            is_accident = predicted_class == 'accident' and confidence >= self.confidence_threshold

            # Store frame result
            frame_result = {
                'frame_number': frame_number,
                'timestamp': frame_number / fps,
                'predicted_class': predicted_class,
                'confidence': float(confidence),
                'is_accident': is_accident
            }
            results['predictions'].append(frame_result)

            # Track accidents
            if is_accident:
                results['summary']['accidents_detected'] += 1
                results['accident_frames'].append(frame_number)
                results['summary']['highest_confidence'] = max(
                    results['summary']['highest_confidence'], confidence
                )

                if accident_start is None:
                    accident_start = frame_number / fps

                # Save frame if requested
                if save_frames:
                    frame_filename = f"accident_frame_{frame_number:06d}.jpg"
                    cv2.imwrite(frame_filename, frame)

            else:
                # End of accident period
                if accident_start is not None:
                    accident_end = frame_number / fps
                    results['summary']['accident_periods'].append({
                        'start_time': accident_start,
                        'end_time': accident_end,
                        'duration': accident_end - accident_start
                    })
                    accident_start = None

            # Draw prediction on frame
            color = (0, 0, 255) if is_accident else (0, 255, 0)
            label = f"{predicted_class}: {confidence:.2%}"
            cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

            if is_accident:
                cv2.putText(frame, "ACCIDENT DETECTED!", (10, 70),
                           cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)
                # Add warning rectangle
                cv2.rectangle(frame, (5, 5), (width-5, height-5), (0, 0, 255), 5)

            # Show frame
            if show_live:
                cv2.imshow('Accident Detection', frame)
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break

            # Save frame to output video
            if output_path:
                out.write(frame)

            frame_number += 1

            # Progress update
            if frame_number % (total_frames // 10) == 0:
                progress = (frame_number / total_frames) * 100
                print(f" Progress: {progress:.1f}%")

        # Cleanup
        cap.release()
        if output_path:
            out.release()
        if show_live:
            cv2.destroyAllWindows()

        # Final summary
        print(f"\n Video Analysis Complete!")
        print(f"   Total frames: {total_frames}")
        print(f"   Accident frames: {len(results['accident_frames'])}")
        print(f"   Accident periods: {len(results['summary']['accident_periods'])}")
        if results['summary']['highest_confidence'] > 0:
            print(f"   Highest confidence: {results['summary']['highest_confidence']:.2%}")

        return results

    def real_time_detection(self, source=0, save_video=False, output_path='realtime_detection.mp4'):
        """
        Real-time accident detection from camera or video stream

        Args:
            source: Camera index (0 for webcam) or video file path
            save_video (bool): Save the detection video
            output_path (str): Path to save output video
        """
        print(f" Starting real-time detection from source: {source}")
        print("Press 'q' to quit, 's' to save screenshot, 'r' to reset stats")

        # Open video source
        cap = cv2.VideoCapture(source)
        if not cap.isOpened():
            raise ValueError(f"Could not open video source: {source}")

        # Get properties
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = int(cap.get(cv2.CAP_PROP_FPS)) if isinstance(source, str) else 30

        # Setup video writer
        if save_video:
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        # Initialize stats
        stats = {
            'total_frames': 0,
            'accidents_detected': 0,
            'start_time': time.time(),
            'last_accident_time': None
        }

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            stats['total_frames'] += 1

            # Make prediction
            processed_frame = self.preprocess_image(frame, 'array')
            start_pred_time = time.time()
            prediction = self.model.predict(processed_frame, verbose=0)
            pred_time = time.time() - start_pred_time

            predicted_class_idx = np.argmax(prediction[0])
            confidence = prediction[0][predicted_class_idx]
            predicted_class = self.class_names[predicted_class_idx]
            is_accident = predicted_class == 'accident' and confidence >= self.confidence_threshold

            # Update stats
            if is_accident:
                stats['accidents_detected'] += 1
                stats['last_accident_time'] = time.time()

            # Draw information on frame
            color = (0, 0, 255) if is_accident else (0, 255, 0)

            # Main prediction
            cv2.putText(frame, f"{predicted_class}: {confidence:.2%}",
                       (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

            # Accident alert
            if is_accident:
                cv2.putText(frame, "ACCIDENT DETECTED!", (10, 70),
                           cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)
                cv2.rectangle(frame, (5, 5), (width-5, height-5), (0, 0, 255), 3)

            # Stats
            runtime = time.time() - stats['start_time']
            fps_current = stats['total_frames'] / runtime if runtime > 0 else 0

            cv2.putText(frame, f"FPS: {fps_current:.1f}", (10, height-90),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            cv2.putText(frame, f"Frames: {stats['total_frames']}", (10, height-60),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            cv2.putText(frame, f"Accidents: {stats['accidents_detected']}", (10, height-30),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            cv2.putText(frame, f"Pred Time: {pred_time*1000:.1f}ms", (width-200, height-30),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

            # Show frame
            cv2.imshow('Real-time Accident Detection', frame)

            # Save video
            if save_video:
                out.write(frame)

            # Handle key presses
            key = cv2.waitKey(1) & 0xFF
            if key == ord('q'):
                break
            elif key == ord('s'):
                # Save screenshot
                screenshot_name = f"screenshot_{datetime.now().strftime('%Y%m%d_%H%M%S')}.jpg"
                cv2.imwrite(screenshot_name, frame)
                print(f" Screenshot saved: {screenshot_name}")
            elif key == ord('r'):
                # Reset stats
                stats = {
                    'total_frames': 0,
                    'accidents_detected': 0,
                    'start_time': time.time(),
                    'last_accident_time': None
                }
                print(" Stats reset")

        # Cleanup
        cap.release()
        if save_video:
            out.release()
        cv2.destroyAllWindows()

        print(f"\n Real-time Detection Summary:")
        print(f"   Total frames processed: {stats['total_frames']}")
        print(f"   Total accidents detected: {stats['accidents_detected']}")
        print(f"   Average FPS: {stats['total_frames']/(time.time()-stats['start_time']):.1f}")

    def _display_single_prediction(self, image_path, result):
        """Display single prediction result"""
        plt.figure(figsize=(12, 6))

        # Load and display image
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(1, 2, 1)
        plt.imshow(img)
        plt.title(f"Input Image\n{Path(image_path).name}")
        plt.axis('off')

        # Display prediction
        plt.subplot(1, 2, 2)
        classes = list(result['all_probabilities'].keys())
        probabilities = list(result['all_probabilities'].values())

        colors = ['red' if cls == 'accident' else 'green' for cls in classes]
        bars = plt.bar(classes, probabilities, color=colors, alpha=0.7)

        # Highlight predicted class
        predicted_idx = classes.index(result['predicted_class'])
        bars[predicted_idx].set_alpha(1.0)
        bars[predicted_idx].set_edgecolor('black')
        bars[predicted_idx].set_linewidth(3)

        plt.title(f"Prediction Results\n{result['predicted_class'].upper()}")
        plt.ylabel('Probability')
        plt.ylim(0, 1)

        # Add percentage labels
        for i, (cls, prob) in enumerate(zip(classes, probabilities)):
            plt.text(i, prob + 0.02, f"{prob:.1%}", ha='center', fontweight='bold')

        plt.tight_layout()
        plt.show()

        # Print detailed results
        print(f"\n Prediction Results for {Path(image_path).name}:")
        print(f"    Predicted Class: {result['predicted_class'].upper()}")
        print(f"    Confidence: {result['confidence']:.2%}")
        print(f"    Accident Alert: {'YES' if result['is_accident'] else 'NO'}")
        print(f"     Prediction Time: {result['prediction_time']:.3f} seconds")

    def _save_prediction_result(self, result, output_dir):
        """Save single prediction result"""
        os.makedirs(output_dir, exist_ok=True)

        filename = f"prediction_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
        filepath = os.path.join(output_dir, filename)

        with open(filepath, 'w') as f:
            json.dump(result, f, indent=2)

        print(f" Result saved to: {filepath}")

    def _save_batch_results(self, results, output_dir):
        """Save batch prediction results"""
        os.makedirs(output_dir, exist_ok=True)

        # Save as JSON
        json_filename = f"batch_results_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
        json_filepath = os.path.join(output_dir, json_filename)

        with open(json_filepath, 'w') as f:
            json.dump(results, f, indent=2)

        # Save as CSV
        csv_filename = f"batch_results_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        csv_filepath = os.path.join(output_dir, csv_filename)

        df = pd.DataFrame(results)
        df.to_csv(csv_filepath, index=False)

        print(f" Batch results saved:")
        print(f"   JSON: {json_filepath}")
        print(f"   CSV: {csv_filepath}")

# Example usage and testing functions
def test_predictor():
    """Test the predictor with sample predictions"""
    try:
        # Initialize predictor
        predictor = AccidentPredictor(
            model_path='final_accident_detection_model.h5',  # Update with your model path
            confidence_threshold=0.7
        )

        print(" Testing Accident Predictor...")

        # Test single image prediction (update with your image path)
        # result = predictor.predict_single_image('test_image.jpg')

        # Test batch prediction (update with your folder path)
        # batch_results = predictor.predict_batch_images('test_images_folder/')

        # Test real-time detection
        # predictor.real_time_detection(source=0)  # Webcam

        print(" Predictor testing completed!")

    except Exception as e:
        print(f" Error testing predictor: {e}")

if __name__ == "__main__":
    # Example usage

    # 1. Single image prediction
    predictor = AccidentPredictor('final_accident_detection_model.h5')

    # Predict single image
    # result = predictor.predict_single_image('path/to/your/image.jpg', show_result=True)

    # Predict batch of images
    # batch_results = predictor.predict_batch_images('path/to/image/folder/')

    # Analyze video
    # video_results = predictor.predict_video('path/to/video.mp4', show_live=True)

    # Real-time detection from webcam
    # predictor.real_time_detection(source=0, save_video=True)

    print(" Prediction system ready! Uncomment the examples above to use.")
```
